```python
def missing_values(df):
    
    total_missing = df.isnull().sum()/df.shape[0]
    percent_missing = total_missing*100
    return percent_missing.sort_values(ascending=False).round(1)
```

```python
num_train_mis_values_before = missing_values(num_df)
```
