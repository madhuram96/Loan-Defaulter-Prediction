```python
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
print ('Ran')
```

```python
merged_train =pd.read_csv('C:\\home-credit-default-risk\\imputed_train.csv')
```

```python
merged_train.head()
```

```python
merged_test =pd.read_csv('C:\\home-credit-default-risk\\imputed_test.csv')
```

```python
merged_test.head()
```

```python
test=merged_test.drop(['SK_ID_CURR'], axis=1)
test.head()
```

```python
X = merged_train.iloc[:,0:464]
y = merged_train[['TARGET']]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2019)
```

```python
models = [
    
    RandomForestClassifier(random_state=2019, n_jobs=-1, 
                           n_estimators=100, max_depth=3),
        
    DecisionTreeClassifier(random_state=2019, 
               max_depth=3)
]
```

```python
from vecstack import stacking
S_train, S_test = stacking(models,                   
                           X, y, test,   
                           regression=False,
     
                           mode='oof_pred_bag', 
       
                           needs_proba=False,
         
                           save_dir=None, 
            
                           metric=accuracy_score, 
    
                           n_folds=1, 
                 
                           stratified=True,
            
                           shuffle=True,  
                           random_state=2019,    
         
                           verbose=1)
```

```python
model = XGBClassifier(random_state=2019, n_jobs=-1, learning_rate=0.1, 
                      n_estimators=50, max_depth=3
```

```python
model = model.fit(S_train, X)
```

```python
S_train.shape
```

```python
y_pred = model.predict(S_test)
```

```python
y_pred_prob = model.predict_proba(S_test)[:,1]
y_pred_prob
```

```python
print('Final prediction score: [%.8f]' % accuracy_score(y_test, y_pred))
```

```python
from sklearn.metrics import roc_curve, roc_auc_score
roc_auc_score(y, y_pred_prob)
```

```python
submission = pd.DataFrame({'SK_ID_CURR': merged_test['SK_ID_CURR'], 'TARGET': y_pred})
```

```python
submission.to_csv('C:\\home-credit-default-risk\\result\\voting_classifier_2.csv', index = False)
```
